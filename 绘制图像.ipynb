**绘制训练结果的图像**

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_data = r"E:\education\road_snow\yolov11m\runs\detect\snow_v5_largefocus6\results.csv"

# 去掉占位符行 “...” 并读取数据
csv_data_cleaned = "\n".join([line for line in csv_data.strip().splitlines() if "..." not in line])
df = pd.read_csv(csv_data_cleaned)

# 提取关键指标绘图
metrics = ['metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)', 'metrics/mAP50-95(B)']

# 绘图：精度、召回率、mAP
plt.figure(figsize=(12, 6))
for m in metrics:
    plt.plot(df['epoch'], df[m], label=m)
plt.title("Precision / Recall / mAP vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Metric")
plt.legend()
plt.grid(True)
plt.tight_layout()

# 保存图片
output_path = "./metrics_plot.png"
plt.savefig(output_path)
```
